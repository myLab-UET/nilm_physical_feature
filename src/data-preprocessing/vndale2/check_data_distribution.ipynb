```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Comparing between 2 combinations

```python
comb_name1 = "led_mayep_maysay_saclt_tulanh.csv"
comb_name2 = "maysay_quat_saclt_tulanh.csv"
BASE_PATH = "/opt/nilm-shared-data/nilm_device_detection/VNDALE_v2/RMS_data/window_1800"
comb1_df = pd.read_csv(f"{BASE_PATH}/test/{comb_name1}")
comb2_df = pd.read_csv(f"{BASE_PATH}/train/{comb_name2}")
print(len(comb1_df), len(comb2_df))
```

```python
columns_to_plot = ["Irms", "MeanPF", "P", "Q", "S"]
fig, axes = plt.subplots(len(columns_to_plot), 2, figsize=(15, 30))
fig.suptitle(f'Distribution of Columns between {comb_name1} and {comb_name2}', fontsize=16)

for i, column in enumerate(columns_to_plot):
    min_val = min(comb1_df[column].min(), comb2_df[column].min())
    max_val = max(comb1_df[column].max(), comb2_df[column].max())
    
    comb1_df[column].hist(ax=axes[i, 0], bins=100, alpha=0.5, color='blue')
    axes[i, 0].set_title(f'{column} - {comb_name1}')
    axes[i, 0].set_xlim(min_val, max_val)
    
    comb2_df[column].hist(ax=axes[i, 1], bins=100, alpha=0.5, color='green')
    axes[i, 1].set_title(f'{column} - {comb_name2}')
    axes[i, 1].set_xlim(min_val, max_val)
    
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
